```python
import subprocess
import ffmpeg
```

```python
def trim_video(input_path, output_path, start, duration):
    cmd = [
        "ffmpeg",
        "-ss", start,
        "-accurate_seek",
        "-i", input_path,
        "-t", duration,
        "-c:v", "libx264",
        "-crf", "23",
        "-c:a", "aac",
        output_path,
        "-y"
    ]
    subprocess.run(cmd, check=True)
```

```python
video_folder = "E:/Data/Overview_trimmed/"
annotations_folder = "E:/Data/ELAN/annotations/"
```

```python

participant_label = {"L": "L", "R": "Rm"}

for session in range(1,3):
    print(session)
    annotation_path = annotations_folder + str(session) + ".txt"
    with open(annotation_path, "r") as a:
        annotations = a.read()
    for participant in participant_label.keys():
        print(participant)
        video = video_folder + str(session) + participant_label[participant] + ".mp4"
        annotation_list = annotations.split("\n")
        for a in annotation_list:
            a = a.split('\t')
            if len(a) == 1:
                continue
            tier, start, duration, label = a[0], a[2], a[6], a[8]
            if tier == f"PUOH-{participant}":
                label = "E:/Data/Clips/PUOH/" + label + ".mp4"
            elif tier == f"Other-{participant}":
                label = "E:/Data/Clips/Other/" + label + ".mp4"
            else:
                continue
            print(duration)
            try:
                trim_video(video, label, start, duration)
            except Exception as e:
                print(f"FAILED: {label}")
                print(f"start={start}, duration={duration}")
                print(f"error={e}")
                continue
```

I used the code below to calculate the total duration of all the video clips

```python
import cv2
import os
import datetime

# Folder containing videos
folder_path = "E:\Data\Clips\PUOH"

# Supported video extensions
video_extensions = ('.mp4', '.avi', '.mov', '.mkv', '.wmv', '.flv')

total_seconds = 0

for file_name in os.listdir(folder_path):
    if file_name.lower().endswith(video_extensions):
        video_path = os.path.join(folder_path, file_name)

        video = cv2.VideoCapture(video_path)

        frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
        fps = video.get(cv2.CAP_PROP_FPS)

        if fps > 0:
            duration_seconds = frames / fps
            total_seconds += duration_seconds

            print(
                f"{file_name}: "
                f"{datetime.timedelta(seconds=round(duration_seconds))}"
            )

        video.release()

# Total duration
total_duration = datetime.timedelta(seconds=round(total_seconds))

print("\n========================")
print(f"Total videos duration: {total_duration}")
print("========================")
```
